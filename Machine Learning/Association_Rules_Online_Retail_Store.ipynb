## Prepare the data to fed into the Apriori Algorithm

```python
france_pivot_data = pd.pivot_table(data = france_data ,values='Quantity',index='InvoiceNo',columns='Description').fillna(0)
france_pivot_data
```

```python
#Custom Function

def encode(x):
    if x<=0:
        return 0
    else:
        return 1
```

```python
france_pivot_data = france_pivot_data.applymap(encode)
france_pivot_data.head(20)
```

```python
#pd.set_option('max_columns',None)
```

## Model Building with Apriori and Association Rules

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
freq_item_sets = apriori(df = france_pivot_data,min_support=0.09,use_colnames=True)
freq_item_sets
```

```python
best_associations = association_rules(df = freq_item_sets,metric = 'confidence',min_threshold = 0.07)
best_associations.sort_values(by = 'support',ascending=False,inplace=True)
```

```python
best_associations
```

```python
best_associations.to_csv('Best Associations.csv')
```
